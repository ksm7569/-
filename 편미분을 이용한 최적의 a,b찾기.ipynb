```python
!pip install numpy

import numpy as np

#데이터를  numpy배열로 저장함(x는 온도,y는 실제 판매량 데이터를 나타냄)
X = np.array([10,7,4,0])

Y = np.array([30,50,60,80])

#a는 초기 기울기, b는 초기 절편
a = -4.0
b = 80.0

#데이터 개수 계싼
m = len(X)

m
```

```python
#반복횟수
epochs = 100

#Gradient Descent : cost를 줄여나가며 최적의 직선을 찾는 방법
for epoch in range(epochs):

    H = a * X+ b
    #오차 = 에츢까ㅃㅆ - 씰쪠값
    error = H -Y

    # 비용은 예측값과 실제값 사이의 차이(오차)를 나타내며,
    # 오차의 크기가 작을수록 우리가 찾은 적선이 실제 데이터를 잘 설명한다는 것을 의미
    cost = (1/m) * np.sum(error ** 2)

    # 기울기(a)의 편미분 값을 계산하여 얼마나 수정해야 할지 정하는 과정입니다.
    # 편미분이란 Cost(오차)를 가장 빨리 줄일 수 있는 방향과 크기를 찾기 위한 계산 방법입니다.
    
    # 계산 방법은 다음과 같습니다.
    # 1. 먼저 예측값(H)과 실제값(Y)의 차이(오차)를 계산합니다. (H - Y)
    # 2. 그 오차를 데이터(X)의 각 값과 곱합니다. 이는 각 데이터가 얼마나 오차에 영향을 미치는지 보기 위함입니다.
    # 3. 위 값을들을 모두 더하고(np.sum), 데이터의 개수(m)로 나눠 평균값을 계산합니다.
    
    # 이 값을 통해, 기울기(a)를 얼마만큼 어느 방향으로 이동시켜야 비용(cost)을 가장 효과적으로 줄일 수 있을지 알 수 있습니다.
    a_grad = (1/m) * np.sum(X * (H - Y))

    # 절편(b)의 편미분 값을 계산하여 얼마나 수정해야 할지를 정하는 과정입니다.
    # 편미분이란, 절편(b)을 아주 조금 변화시킬 때 비용(cost)이 얼마나 바뀌는지를 알아내는 계산 방법입니다.
    
    # 계산 방법은 다음과 같습니다:
    # 1. 예측값(H)과 실제값(Y)의 차이를 계산하여 오차를 구합니다. (H - Y)
    #    오차가 클수록 현재 직선이 데이터를 잘 설명하지 못하는 것입니다.
    # 2. 이 오차를 모두 더한 후(np.sum), 데이터 개수(m)로 나누어서 평균값을 계산합니다.
    
    # 이렇게 얻은 값(b_grad)을 사용하면, 절편(b)을 얼마나 조정해야 오차가 가장 빨리 줄어드는지 알 수 있습니다.
    b_grad = (1/m) * np.sum(H - Y)

    # 이전 단계에서 계산한 편미분 값을 사용하여 기울기(a)를 업데이트(갱신)합니다.
    # a에서 a_grad 값을 빼는 이유는, Gradient Descent(경사 하강법) 원리에 따라
    # a를 오차가 가장 빠르게 줄어드는 방향으로 이동시키기 위해서입니다.
    # a_grad 값은 오차를 줄이기 위한 방향과 크기를 나타냅니다.
    a = a - a_grad
    
    # 이전 단계에서 계산한 편미분 값을 사용하여 절편(b)을 업데이트(갱신)합니다.
    # b에서 b_grad 값을 빼는 이유는, 절편 역시 오차를 가장 빨리 줄일 수 있는 방향으로 이동시키기 위함입니다.
    # b_grad 값은 절편을 얼마나, 어떤 방향으로 바꿔야 할지를 나타냅니다.
    b = b - b_grad

    # 반복이 끝나면 최종적으로 계산된 가장 좋은 기울기와 절편을 출력합니다.
    # 이 직선이 데이터를 가장 잘 나타내는 직선이 됩니다.
    print(f'\n최적화된 선형회귀식: Y = {a:.4f}X + {b:.4f}')

```
